```python
# Import necessary libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Get user inputs
ticker = input("Enter the stock ticker symbol (e.g., 'MSFT'): ")
start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")
```

```python
# Fetch the stock data from Yahoo Finance
stock_data = yf.download(ticker, start=start_date, end=end_date)
```

```python
# Filter only the 'Close' column
stock_data = stock_data[['Close']]
```

```python
# Plot the daily closing prices
plt.figure(figsize=(10, 6))
plt.plot(stock_data.index, stock_data['Close'], marker='o', linestyle='-')
plt.title(f'Daily Closing Prices for {ticker} from {start_date} to {end_date}')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.show()
```

```python
# Calculate the average closing price

average_close = stock_data['Close'].mean()
print(f"The average closing price for {ticker} from {start_date} to {end_date} is: ${average_close:.2f}")
```
